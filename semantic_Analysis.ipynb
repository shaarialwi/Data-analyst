```python
#survey, check sentiment analysis

import pandas as pd
import re
senti = open('OpinionFinder_Lexicon.tff','r').read().splitlines()
sentiment={}
for line in senti:
    tokens = line.split(' ')
    term = tokens[2].replace('word1=','')
    if tokens[0] == 'type=weaksubj':
        score = 1
    elif tokens[0] == 'type=strongsubj':
        score = 3
    if tokens[5] == 'priorpolarity=negative':
        polarity = -1
    elif tokens[5] == 'priorpolarity=positive':
        polarity = 1
    sentiment[term] = polarity * score
print(sentiment)
df = pd.DataFrame.from_dict(sentiment,orient='index')
import csv
with open('sentiment.csv', 'w') as f:  # Just use 'w' mode in 3.x
    w = csv.writer(f)
    w.writerows(sentiment.items())
df.sample(20)
```

```python
def calculate_sentiment(d):
    import re
    tokens = re.findall(r'\w+', d.lower())
    sentiment_score = 0
    for token in tokens:
        if token in sentiment:
            sentiment_score = sentiment_score + sentiment[token]
            print(token, sentiment[token])
    return sentiment_score
```

```python
d = 'A positively thrilling combination of ethnography and all the intrigue , betrayal , deceit and murder of a Shakespearean tragedy or a juicy soap opera .'
sentiment_score = calculate_sentiment(d)
print(f"sentiment_score : {sentiment_score}")
```
